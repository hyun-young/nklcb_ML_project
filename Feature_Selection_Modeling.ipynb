````python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score

models = []
models.append( ('AdaBoostClassifier', AdaBoostClassifier()) )
models.append( ('GradientBoostingClassifier', GradientBoostingClassifier()) )
models.append( ('RandomForestClassifier', RandomForestClassifier()) )
models.append( ('DecisionTreeClassifier', DecisionTreeClassifier()) )
models.append( ('LGBMClassifier', LGBMClassifier()) )
````

````python
for name, model in models:
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    print(f'{name}의 test 데이터 accuracy score : ', accuracy_score(y_test, pred))
````

````python
from sklearn.model_selection import KFold

kfold = KFold(n_splits=5, random_state=13, shuffle=True)
cv_accuracy = []

for train_idx, test_idx in kfold.split(X):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
    cv_acc_dic = {}
    for name, model in models:
        model.fit(X_train, y_train)
        pred = model.predict(X_test)
        cv_acc_dic[name] = accuracy_score(y_test, pred)
        
    cv_accuracy.append(cv_acc_dic)

cv_accuracy_df = pd.DataFrame(cv_accuracy)
cv_accuracy_df
````

````python
# CV Score 평균
cv_mean_std_df= pd.DataFrame([cv_accuracy_df.mean(), cv_accuracy_df.std()], index=['평균', '표준편차']).transpose()
for name, score in cv_mean_std_df.iterrows():
    print(f'{name} CV Score 평균 = {round(score[0],5)}, 표준편차 = {round(score[1],5)}')
````
